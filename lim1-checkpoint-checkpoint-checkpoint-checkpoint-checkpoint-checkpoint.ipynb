```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define input shape and batch size
input_shape = (150, 150, 3)  # Image dimensions and 3 for RGB color channels
batch_size = 32

# Create data generators for training and validation
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

train_generator = train_datagen.flow_from_directory(
    r'C:\Users\preet\OneDrive\Desktop\DataSet\Train',
    target_size=input_shape[:2],
    batch_size=batch_size,
    class_mode='binary'  # Assuming binary classification, change to 'categorical' for multi-class
)

validation_datagen = ImageDataGenerator(rescale=1.0/255.0)

validation_generator = validation_datagen.flow_from_directory(
    r'C:\Users\preet\OneDrive\Desktop\DataSet\Valid',
    target_size=input_shape[:2],
    batch_size=batch_size,
    class_mode='binary'  # Assuming binary classification, change to 'categorical' for multi-class
)

# Build the CNN model
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification

# Compile the model
model.compile(optimizer=Adam(lr=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(
    train_generator,
    epochs=20,
    validation_data=validation_generator
)

# Evaluate the model on the test dataset (assuming you have a separate test generator)
# test_loss, test_accuracy = model.evaluate(test_generator)
# print(f"Test Loss: {test_loss:.4f}")
# print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
# Import necessary libraries
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import preprocess_input
import numpy as np
from lime.lime_image import LimeImageExplainer
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt

# Define the preprocess_img function
def preprocess_img(img_path):
    # Load the image and resize it to the input size expected by your model
    img = image.load_img(img_path, target_size=(150, 150))
    
    # Convert the image to a numpy array
    img_array = image.img_to_array(img)
    
    # Expand the dimensions to create a batch of size 1
    img_array = np.expand_dims(img_array, axis=0)
    
    # Preprocess the image using the same preprocessing applied during model training
    img_preprocessed = preprocess_input(img_array)
    
    return img_preprocessed

# Create a LIME explainer
explainer = LimeImageExplainer()

# Choose an image to explain (replace 'path_to_image.jpg' with the actual image path)
img_path = r'C:\Users\preet\Downloads\WhatsApp Image 2023-10-01 at 2.58.38 AM.jpeg'

# Preprocess the image
img = preprocess_img(img_path)

# Generate explanations using LIME
explanation = explainer.explain_instance(
   img[0],  # LIME expects a 2D array, so use img[0] to get the image data
   lambda x: model.predict(x),  # Wrap model.predict inside a lambda function
   top_labels=1,  # Number of top labels to explain (adjust as needed)
   hide_color=0,  # Color for masked portions of the image (0 corresponds to black)
   num_samples=1000  # Number of random samples to generate
)

# Visualize LIME explanation
temp, mask = explanation.get_image_and_mask(
   explanation.top_labels[0], positive_only=False, num_features=10, hide_rest=False
)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
plt.show()
```
